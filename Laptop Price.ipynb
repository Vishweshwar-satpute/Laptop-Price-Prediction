```python
# Things to do 
# > Remove gb from the Ram column
# > Remove measure from the Weight (Kg)
# > Convert PriceEuros to Indian-Rs
```

```python
# Remove gb from the Ram column
```

```python
gb = lapy['Ram'].str[-2:]
gb
```

```python
lapy['Ram in GB'] = lapy['Ram'].str[:-2]
lapy['Ram in GB'].value_counts()
```

```python
lapy.head()
```

```python
lapy.drop(columns='Ram', inplace =True)                        #removing/drop the Column Ram from the data set as we have created Ram in gb.
```

```python
lapy['Ram in GB'] = pd.to_numeric(lapy['Ram in GB'])   
print(lapy['Ram in GB'])                                       #changing the Ram type from str to int
```

```python
print(lapy.dtypes)
```

```python
lapy.head()            
```

```python
# > Remove measure from the Weight (Kg)
```

```python
lapy[~lapy['Weight'].str.contains('kg') ]                 #to check any values do not contains Kg.
```

```python
lapy['Weights in Kg'] = lapy['Weight'].str.replace('kg','')               #As the measure is common for all the weights so removing Kg.
```

```python
lapy.drop(axis=1, columns= 'Weight', inplace = True)
```

```python
lapy['Weights in Kg'].astype(float)
```

```python
lapy['Weights in Kg'] = pd.to_numeric(lapy['Weights in Kg'])

print (lapy.dtypes)
```

```python
lapy.head()
```

```python
# > Convert PriceEuros to Indian-Rs
# Current price 1 Euro = 77.3199 Indian Rupies

lapy['Price in ₹'] = lapy['Price_euros'] * 77.3199
lapy['Price in ₹']
```

```python
lapy.drop(axis=1, columns= 'Price_euros', inplace= True)
```

```python
lapy.head()
```

```python
lapy['Price in ₹'] = lapy['Price in ₹'].apply(np.floor).astype(int)
lapy['Price in ₹'].to_frame()
```

# Filter

```python
lapy['Company'].value_counts().plot(kind = 'bar', figsize = (10,12),fontsize = 12, grid =True)  # No of laptops per company.
plt.title('Number of laptops per Company')
plt.xlabel('Company Names')
plt.ylabel('Number of Laptops')
```

```python
space = np.ones(6)/10

lapy['TypeName'].value_counts().sort_values().plot(kind= 'pie', shadow = True, figsize=(10,10), fontsize=12, autopct= '%.2f%%', explode = space, wedgeprops=dict(width = 0.18))   # Different types of laptops to offer by the Company.
plt.title('Showing different Type/Category of Laptop')
plt.ylabel("")
```

```python
lapy['Inches'].value_counts().sort_values(ascending = False).plot(kind= 'bar',  figsize=(15,7), cmap='PRGn_r', fontsize=12,grid = True, linestyle='-.')  
plt.ylabel('Count')
plt.xlabel('Screen Inches')
plt.title('Screen-size in Inches of each and every Laptop offered by the Company')
```

```python
space = np.ones(9)/10

lapy['OpSys'].value_counts().sort_values().plot(kind = 'pie', figsize=(20,20), fontsize = 14, explode = space, wedgeprops=dict(width = 0.25),autopct='%.2f%%')
plt.ylabel("")
plt.title("Different Operating System's with their Percentages")
```

# Multi filter
